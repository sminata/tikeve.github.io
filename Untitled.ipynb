```python
def oper(a,b,n):
    if (a=='')|(b=='')|((b==0)&(n==4))|((a==0)&(n==6)):
        return ''
    elif n == 1:
        return a+b
    elif n == 2:
        return a*b
    elif n == 3:
        return a-b
    elif n == 4:
        return a/b
    elif n==5:
        return b-a
    elif n==6:
        return b/a

oper(1,2,4)
```
